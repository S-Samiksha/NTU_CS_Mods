```python
# Basic Libraries
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt # we only need pyplot
sb.set() # set the default Seaborn style for graphics
```

```python
html_data = pd.read_html('http://en.wikipedia.org/wiki/2016_Summer_Olympics_medal_table')
print("Data type : ", type(html_data))
print("HTML tables : ", len(html_data))
html_data[1].head()
```

```python
OlympicsMedalTable = html_data[1]
TOP_20 = OlympicsMedalTable.head(20)
TOP_20.head()
```

```python
year=2016
n=0
TOP_dict = {
    "OlmpicsMedalTable2016": [],
    "OlmpicsMedalTable2012": [],
    "OlmpicsMedalTable2008": [],
    "OlmpicsMedalTable2004": [],
    "OlmpicsMedalTable2000": []
}

while(n<4):
    html_data = pd.read_html('http://en.wikipedia.org/wiki/'+str(year)+'_Summer_Olympics_medal_table')
    TOP_dict[n] = html_data[1].head(20)
    year = year-4
    n+=1
```

```python
TOP_dict[3].head()
```

```python
NOC2016 = pd.DataFrame(TOP_dict[0]['NOC'])
Rank2016 = pd.DataFrame(TOP_dict[0]['Rank'])
joint2016 = pd.concat([NOC2016, Rank2016], axis = 1, names=['NOC', '2016']).reindex(NOC2016.index)
```

```python
NOC2012 = pd.DataFrame(TOP_dict[1]['NOC'])
Rank2012 = pd.DataFrame(TOP_dict[1]['Rank'])
joint2012 = pd.concat([NOC2012, Rank2012], axis = 1).reindex(NOC2012.index)
```

```python
NOC2008 = pd.DataFrame(TOP_dict[2]['NOC'])
Rank2008 = pd.DataFrame(TOP_dict[2]['Rank'])
joint2008 = pd.concat([NOC2008, Rank2008], axis = 1).reindex(NOC2008.index)
NOC2004= pd.DataFrame(TOP_dict[3]['Nation'])
Rank2004 = pd.DataFrame(TOP_dict[3]['Rank'])
joint2004 = pd.concat([NOC2004, Rank2004], axis = 1).reindex(NOC2004.index)
```

```python
final_table=pd.merge(joint2016, joint2012, on="NOC", suffixes=('_2016', '_2012')) 
```
