```python
year=2016
n=0
TOP_dict = {
    "OlmpicsMedalTable2016": [],
    "OlmpicsMedalTable2012": [],
    "OlmpicsMedalTable2008": [],
    "OlmpicsMedalTable2004": [],
    "OlmpicsMedalTable2000": []
}

while(n<5):
    html_data = pd.read_html('http://en.wikipedia.org/wiki/'+str(year)+'_Summer_Olympics_medal_table')
    TOP_dict[n] = html_data[1].head(30)
    year = year-4
    n+=1
```

```python
for i in range(0,5):
    TOP_dict[i].columns = ['Rank', 'Nation', 'Gold', 'Silver', 'Bronze', 'Total']
    TOP_dict[i].drop(['Total', 'Gold', 'Silver', 'Bronze'], axis=1, inplace=True)
```

```python
final_data = pd.merge(TOP_dict[0], TOP_dict[1], on="Nation").merge(TOP_dict[2], on="Nation").merge(TOP_dict[3], on="Nation").merge(TOP_dict[4], on="Nation")
```

```python
final_data.columns = ['2016', 'Nation', '2012', '2008', '2004', '2000']
final_data.head(20)
```
